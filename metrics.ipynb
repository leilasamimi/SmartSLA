```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# داده‌های Table 4
data = {
    'Metamodel': ['IcontractML', 'IcontractML2.0', 'BitML', 'AsyncSLA', 'SmartSLA'],
    'Maintainability': [10.4, 12.6, 25.6, 9.4, 18.4],
    'Understandability': [0.66, 0.45, 1.143, 1.235, 1.0625],
    'Extensibility': [0.1, 0.08, 0.0625, 0.238, 0.0161]
}

df = pd.DataFrame(data)

# تنظیمات اولیه
labels = ['Maintainability', 'Understandability', 'Extensibility']
num_vars = len(labels)

# تبدیل زاویه‌ها به مختصات قطبی
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # بستن دایره

# رسم نمودار
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# حلقه رسم برای هر متامدل
for i in range(len(df)):
    values = df.iloc[i][labels].tolist()
    values += values[:1]  # بستن دایره
    ax.plot(angles, values, label=df['Metamodel'][i], linewidth=2)
    ax.fill(angles, values, alpha=0.1)

# تنظیمات نمایشی
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)
plt.title('Radar Chart of Structural Metrics for SLA Metamodels', size=14)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# مقادیر خام
maint_raw = [10.4, 12.6, 25.6, 9.4, 15.8]
understand_raw = [0.66, 0.45, 1.143, 1.235, 1.625]
extens_raw = [0.1, 0.08, 0.0625, 0.238, 0.14]

# معکوس کردن maintainability چون کمتر بودن بهتر است
maint_flipped = [max(maint_raw) - x for x in maint_raw]

# تابع نرمال‌سازی به بازه 0 تا 1
def normalize(data):
    min_val = min(data)
    max_val = max(data)
    return [(x - min_val) / (max_val - min_val) for x in data]

# نرمال‌سازی هر معیار
maint_norm = normalize(maint_flipped)
understand_norm = normalize(understand_raw)
extens_norm = normalize(extens_raw)

# ساخت دیتافریم
data = {
    'Metamodel': ['iContractML', 'iContractML 2.0', 'BitML', 'AsyncSLA', 'SmartSLA'],
    'Maintainability': maint_norm,
    'Understandability': understand_norm,
    'Extensibility': extens_norm
}

# تبدیل به فرمت مناسب برای seaborn
df = pd.DataFrame(data)
df_melted = df.melt(id_vars='Metamodel',
                    value_vars=['Maintainability', 'Understandability', 'Extensibility'],
                    var_name='Metric', value_name='Value')

# رسم نمودار
plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Metric', y='Value', hue='Metamodel')
plt.title('Normalized Comparison of Structural Quality Metrics across SLA Metamodels')
plt.ylabel('Normalized Metric Value (0–1, Higher is Better)')
plt.xlabel('Structural Metric')
plt.ylim(0, 1.05)
plt.legend(title='Metamodel', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
